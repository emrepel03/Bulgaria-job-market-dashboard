**Top Skills in Demand**

```python
df_skills = pd.read_csv("jobs_per_skill.csv")
top_skills = df_skills.sort_values(by="job_count", ascending=False).head(20)

plt.figure(figsize=(10, 6))
sns.barplot(x="job_count", y="skill", data=top_skills, palette="viridis")
plt.title("Top 20 In-Demand Skills in Bulgaria Tech Jobs")
plt.xlabel("Number of Job Postings")
plt.ylabel("Skill")
plt.tight_layout()
plt.show()
```

**Top Hiring Companies**

```python
df_companies = pd.read_csv("jobs_per_company.csv")
top_companies = df_companies.sort_values(by="job_count", ascending=False).head(15)

plt.figure(figsize=(10, 6))
sns.barplot(x="job_count", y="company", data=top_companies, palette="cubehelix")
plt.title("Top 15 Hiring Companies in Bulgaria's Tech Sector")
plt.xlabel("Number of Job Postings")
plt.ylabel("Company")
plt.tight_layout()
plt.show()
```

**Job Distribution by Seniority Level**

```python
df_levels = pd.read_csv("jobs_per_level.csv")
df_levels = df_levels[df_levels["level_group"] != "Unspecified"]

order = ["Intern", "Junior", "Mid", "Senior", "Lead"]
df_levels["level_group"] = pd.Categorical(df_levels["level_group"], categories=order, ordered=True)
df_levels = df_levels.sort_values("level_group")

plt.figure(figsize=(8, 5))
sns.barplot(x="job_count", y="level_group", data=df_levels, palette=sns.color_palette("hls", len(df_levels)))
plt.title("Job Distribution by Experience Level")
plt.xlabel("Number of Job Postings")
plt.ylabel("Experience Level")
plt.tight_layout()
plt.show()
```

**Job Count by Position Keyword**

```python
df_positions = pd.read_csv("jobs_per_position_keyword.csv")

df_positions = df_positions.sort_values(by="job_count", ascending=True)

palette = sns.color_palette("hls", len(df_positions))

plt.figure(figsize=(10, 6))
sns.barplot(
    x="job_count",
    y="position",
    data=df_positions,
    palette=palette
)

plt.title("Job Count by Position Keyword")
plt.xlabel("Number of Jobs")
plt.ylabel("Position Keyword")
plt.tight_layout()
plt.show()
```

**Jobs by Location**

```python
df_locations = pd.read_csv("jobs_per_location.csv")
df_locations = df_locations.sort_values("job_count", ascending=True)
df_locations["percent"] = (df_locations["job_count"] / df_locations["job_count"].sum() * 100).round(1)

plt.figure(figsize=(8, 5))
bars = sns.barplot(data=df_locations, x="job_count", y="location", palette="Set2")

for i, (count, pct) in enumerate(zip(df_locations["job_count"], df_locations["percent"])):
    bars.text(count + 3, i, f"{count} ({pct}%)", va='center')

plt.title("Job Distribution by City")
plt.xlabel("Number of Job Postings")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

Seniority Distribution per Role

```python
df_top_roles_vs_levels = pd.read_csv("top_roles_vs_levels.csv")
df_melted = df_top_roles_vs_levels.melt(id_vars="role", 
                                        var_name="seniority", 
                                        value_name="job_count")

plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x="role", y="job_count", hue="seniority", palette="coolwarm")
plt.title("Seniority Distribution per Role")
plt.xlabel("Role")
plt.ylabel("Job Count")
plt.legend(title="Seniority")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
